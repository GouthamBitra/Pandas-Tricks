# Merging and concat

```python
d1=pd.DataFrame([['a',1],['b',2]],columns=['col1','number'])
d2=pd.DataFrame([['c',3,'Lion'],['d',4,'Tiger']],columns=['Letter','number','animal'])
```

```python
#columnwise
pd.concat([d1,d2], axis=0, ignore_index=True)
```

```python
#Row wise
pd.concat([d1,d2],axis=1)
```

```python
d1=pd.DataFrame({
    "City":["Bangalore","Tirupati","Chennai","Goa"],
    "Temp":[34,40,45,31]
    
})
```

```python
d2=pd.DataFrame({
    "City":["Tirupati","Banaglore","Goa"],
    "Humidity":[68,70,90]
})
```

```python
df=pd.merge(d1,d2,on='City')
df
```
